````python
def calculate_variance(nl):
    sum_list = 0
    for n in nl:
        sum_list = sum_list + n
    mean = sum_list / len(nl)
    sum_squares = 0
    for n in nl:
        sum_squares = sum_squares + (n - mean)**2
    variance = sum_squares / len(nl)
    return variance
````
